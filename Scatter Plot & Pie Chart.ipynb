# Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_cords = np.random.random(200) * 100
y_cords = np.random.random(200) * 100
```

```python
plt.scatter(x_cords,y_cords,edgecolor='red')
plt.show()
```

# Pie Chart

```python
population = [121,35,5,140,14]
countries = ['India','England','Nepal','China','Dubai']
```

```python
explode = [0,0,.5,.3,0] # Exploding countries having maximu and minimum population
colors = ['b','g','r','y','m']
plt.figure(figsize=(8,10))
plt.pie(population,labels = countries, explode = explode, colors = colors)
plt.title("Countries's Population")
plt.legend()
plt.show()
```
