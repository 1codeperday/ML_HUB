```python
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x1 = list(range(0,20))
x2 = list(range(0,15))
```

```python
y1 = np.random.random(20)
y2 = np.random.random(15)
```

```python
plt.plot(x1,y1,'b--',marker='o',label = "Blue")
plt.plot(x2,y2,'g-.',marker='o',label = "Green")
plt.legend(loc='upper right')
plt.show()
```

# Bar Graph

```python
id = [1,2,3,4,5]
```

```python
marks = [85,65,75,95,70]
```

```python
plt.bar(id,marks,color='blue')
# mixing line & bar graph
plt.plot(id,marks)

plt.title("Bar-Graph Sample")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
