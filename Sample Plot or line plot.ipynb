```python
import matplotlib
import matplotlib.style as style
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = list(np.arange(1,21))
x
```

```python
y = np.random.random(20) 
y
```

```python
plt.plot(x,y,'b--',marker='o')
'''
If we want to change the color of graph line then we have to pass kayword argument as color ='blue' OR it is third 
parameter which accepts string.Apart from color we can also pass dash --
If we want to add some markers in the points where we are generating these values 
for this we have an additional parameter by the name of marker.
'''
plt.grid()
# If we want to add grid in our graph we can do it by using plt.grid()
plt.show()
```

```python
'''
If we wish to use style then we need to import style() module from matplotlib egs.
>>>import matplotlib.style as style

Then we need call use() function and pass its argument.egs.
>>>style.use('ggplot')
'''

style.use('ggplot')
plt.plot(x,y,'b--',marker='o')
plt.title('Sample Graph')

# In order to mention axies we use plt.xlabel('X-axis') & plt.ylabel('Y-axis')
# plt.xlabel('X-axis')

# To save the figure we need to call plt.savefig() function which accepts file name (mandatory) and path(optinal , by default where you are writing code.)
# plt.savefig('C:\\Users\\HP\\ML_HUB\\Matplotlib\\Sample1')
# Note: - we need to call plt.savefig('File_Name') inside that shell where we call plt.plot() function.
plt.savefig('C:\\Users\\HP\\ML_HUB\\Matplotlib\\Sample1')

# If we wish to modify the interval values of x-cordinates and y-coordinates, we can do it by using plt.xticks() and plt.yticks() function which accepts list as
plt.xticks([0,5,10,20])
plt.yticks([0,.2,.4,.6,.8,1])
plt.show()
```
