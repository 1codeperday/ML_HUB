```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## Subplot

```python
x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([0, 1 ,2, 3, 4])
```

```python
y1 = np.array([10, 20, 30, 40, 50])
y2 = np.random.random(5)
y2 = y2*100 #Broadcasting
```

```python
left = 0.125  
right = 0.9 
bottom = 0.1
top = 0.9    
wspace = 0.5
hspace = 0.6

plt.figure(figsize=(10,8))

plt.subplots_adjust(left, bottom, right, top, wspace, hspace)

plt.subplot(2,3,6)
plt.plot(x1,y1)

plt.subplot(2,3,1)
plt.bar(x2,y2)

x1 = x1*2
y1 = y1*17

x2 = x2*100

plt.subplot(2,3,3)
plt.bar(x1,y1)
plt.title("Our graph with position 3")
plt.show()
```

```python
plt.subplot(5,3,15)
plt.subplot(5,3,1)
plt.subplot(5,3,6)
plt.subplot(5,3,3)
plt.subplot(5,3,7)
plt.subplot(5,3,11)
plt.bar(x1,y1)
plt.title("Our graph with position 3")
plt.show()
```
